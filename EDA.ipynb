```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: better plot style
sns.set(style="whitegrid")
```

```python
# Example
df = pd.read_csv('C:\\Users\\junji\\OneDrive\\Desktop\\stock_data.csv')  # <-- replace with your filename if needed
df.head()
```

```python
# Show data info
print(df.info())

# Show basic statistics
print(df.describe())

# Value counts for each categorical feature
for col in df.select_dtypes(include=['object', 'category']).columns:
    print(f"\nValue Counts for {col}:")
    print(df[col].value_counts())
```

```python
sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12,8))
# Select only numeric columns
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.hist(figsize=(15,10), bins=30, edgecolor='black')
plt.suptitle('Histograms of Numerical Features', y=1.02)
plt.show()
```

```python
# Draw boxplots for numerical columns
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(6,4))
sns.scatterplot(x=num_cols[0], y=num_cols[1], data=df)
plt.title(f'Scatter Plot between {num_cols[0]} and {num_cols[1]}')
plt.show()
```

```python
### Observations:
- Pairplot: Feature X and Y seem positively correlated.
- Heatmap: Feature Z is highly negatively correlated with Target.
- Histograms: Feature A is right-skewed.
- Boxplots: Outliers detected in Feature B and Feature C.
- Scatterplot: Nonlinear relationship observed between Feature1 and Feature2.
```
